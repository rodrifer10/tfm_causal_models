# Separación Train-Test Stratified

Separaré el dataset en Train y Test manteniendo la proporción de la variable objetivo. Pero antes, graficaré la proporción actual de dicha variable.

```python
target_count = df_fraud.groupby('Class').agg({'Class':'count'}).reset_index(drop=True)
target_count['value'] = list(target_count.index)

fig, ax = plt.subplots()
sns.barplot(data=target_count, x='value', y='Class', ax=ax, hue='Class',legend=False, palette=['crimson','deepskyblue'])
ax.set_title('Conteo de valores de la variable Class', fontdict={'fontsize':15}) #,'fontname':'Montserrat','weight':'bold'
ax.yaxis.set_major_formatter('{x:,.0f}')
for i in ax.containers:
    ax.bar_label(i,fmt='{:,.0f}')
ax.set_ylabel('Count')
ax.set_xlabel('Value')
plt.show()
```

```python
target_perc = df_fraud.Class.value_counts(normalize=True).reset_index()

fig, ax = plt.subplots()
sns.barplot(data=target_perc, x='Class', y='proportion', ax=ax, hue='proportion',legend=False, palette=['crimson','deepskyblue'])
ax.set_title('Conteo de valores de la variable Class', fontdict={'fontsize':15}) #,'fontname':'Montserrat','weight':'bold'
#ax.yaxis.set_major_formatter('{x:,.0f}')
for i in ax.containers:
    ax.bar_label(i,fmt='{:.2%}')
plt.yscale('log')
ax.set_ylabel('Proportion')
ax.set_xlabel('Value')
plt.show()
```

```python
target_count
```

```python
df_fraud.Class.value_counts(normalize=True).reset_index()
```

Calculé y grafiqué los valores de la variable Target, no solo para volver a mostrar su distribución, sino también para combrobar que luego de
llevar a cabo el split en train y test las proporciones se mantengan gracias a la estratificación.

```python
X_df_fraud_train, X_df_fraud_test, y_df_fraud_train, y_df_fraud_test = train_test_split(df_fraud.drop('Class',axis=1)
                                                                         ,df_fraud['Class']
                                                                         ,stratify=df_fraud['Class']
                                                                         ,test_size=0.3
                                                                         ,random_state=seed)
df_fraud_train = pd.concat([X_df_fraud_train, y_df_fraud_train],axis=1)
df_fraud_test = pd.concat([X_df_fraud_test, y_df_fraud_test],axis=1)
```

```python
print(f'''
\033[1mTRAIN\033[0m:
{y_df_fraud_train.value_counts(normalize=True)}

\033[1mTEST\033[0m:
{y_df_fraud_test.value_counts(normalize=True)}''')
```

La estratificación fue un éxito, tenemos la misma proporción tanto en Train como en Test

# Detalles a tener en cuenta en base al análisis gráfico

Es interesante ver como algunas variables cambian su comportamiento en relación a la variable objetivo. Algunas anotaciones a tener en cuenta:

1. Hay variables que parecen modificar levemente su comportamiento cuando se produce el fraude, como pueden ser las V1, V2, V9 y V18
2. Mientras tanto hay otras variables que modifican su comportamiento de manera aún más notoria cuando se produce el fraude, siendo su mediana y distribución muy diferente a los registros habituales. Estas variables son las V3, V4, V10, V11, V12, V14, V16 y V17
3. Las demás variables no muestran comportamientos tan diferentes ante la aparición o no de fraude en el target.
4. Sorprendentemente el monto de la transacción NO parece ser determinante en cuanto a la existencia o no de fraude, por un lado seguramente debido a que la gran mayoría de las transacciones parecen ubicarse dentro del mismo rango de pequeñas sumas de dinero, pero aún así, la distribución de la variable en fraude y no fraude no parece mostrar una diferencia significativa.
5. En cuanto a las distribuciones de las variables, todas parecen tener una distribución que podría asemejarse a una leptocúrtica más que a la normal en general, ya que se puede observar alta curtosis en la mayoría de las variables analizadas. También hay otras que se asemejan a la normal y unas pocas que parecen ser bimodales (por ejemplo la variable Time o las V4, V22 o V26) o con algunas modas más aún (V1 o V24).

No hay que perder de vista que la variable objetivo está muy desbalanceada, son muy pocos los registros con fraude (cerca del 0.17% del total de registros). Es posible que, antes de ejecutar el modelado, se requiera un procesamiento de oversampling o undersampling para compensar un poco dicho desbalance.

## Análisis de Relaciones entre las variables

### Correlation Matrix para variables numéricas


```python
corr = df_fraud_train.select_dtypes('number').corr(method='pearson')
corr
```

```python
fr.plot_corr(corr, target_text=True) # función propia importada, revisar en el .py de funciones en caso de querer entender su funcionamiento
```

Muestro la matriz de correlaciones, en donde solo destacan relaciones altas entre unas pocas variables. Comentaré las mismas luego de un pequeño análisis extra individual:

```python
new_corr = corr
new_corr.loc[:,:] = np.tril(new_corr, k=-1) # mantengo sólo una parte de las dos
new_corr = new_corr.stack().to_frame('correlation').reset_index().sort_values(by='correlation', ascending=False)
new_corr[(new_corr['correlation']>0.4) | (new_corr['correlation']<-0.4)].style.highlight_max('correlation',color='mistyrose').highlight_min('correlation',color='lightcyan')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df_fraud_train[df_fraud_train.Class==0],x='V2',y='Amount', color='deepskyblue', alpha=0.5, label='No Fraud')
sns.scatterplot(data=df_fraud_train[df_fraud_train.Class==1],x='V2',y='Amount', color=['Crimson'], alpha=1, label='Fraud')
plt.ylim(-100,5000)
plt.xlim(-50,25)
plt.title('Scatterplot de la variable V2 vs Amount', fontdict={'fontsize':15})
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df_fraud_train[df_fraud_train.Class==0],x='V7',y='Amount', color='deepskyblue', alpha=0.5, label='No Fraud')
sns.scatterplot(data=df_fraud_train[df_fraud_train.Class==1],x='V7',y='Amount', color=['Crimson'], alpha=1, label='Fraud')
plt.ylim(-100,5000)
plt.xlim(-45,30)
plt.title('Scatterplot de la variable V7 vs Amount', fontdict={'fontsize':15})
plt.show()
```

De las variables analizadas podemos observar cierta correlación:

1. Las variables V2 y Amount parecen tener cierta correlación negativa, es decir, que mientras más chico es el amount más grande parece ser V2 y viceversa. Es interesante ver como los fraudes en este caso se centran en un V2 mayor a 0 y un amount chico. De igual manera, la correlación no es tan alta como para implicar la necesidad de desacernos de una de las dos variables.

2. Por otro lado, vemos una correlación positiva entre V7 y Amount, es decir que, al contrario que el caso anterior, V7 parece aumentar junto con el aumento del monto de la transacción.

Además de estas correlaciones, no hay muchas más que destaquen, además de que ninguna llega a valores de correlación altos como para suponer la unión o eliminación de variables.

```python
pd.DataFrame(corr.loc['Class'].sort_values(ascending=False)).style.background_gradient(cmap='RdBu_r', vmin=-0.3, vmax=0.3)
```

Es muy interesante ver como ciertas variables están más correlacionadas con el target, especialmente las V17, V14 y V12.
