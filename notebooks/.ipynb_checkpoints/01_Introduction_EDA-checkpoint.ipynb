### Acerca del DataSet:
Se trata de una base de datos transacciones con tarjetas de crédito, en donde cada instancia representa una transacción. Los datos no solo están completamente anonimizados, sino que conceptualmente es imposible diferenciar entre la gran mayoría de las variables, ya que tanto sus nombres están encodificados como sus valores también, debido a que se ha realizado una transformación mediante el método de PCA (Principal Components Analysis). Esto en principio no debería alterar el desarrollo del modelado, pero sin dudas limita la interpretación desde el punto de vista conceptual del caso, ya que, aunque podamos detectar las principales variables que influyen en el caso, no sabremos con exactitud a que concepto real hace referencia cada una de ellas.

Las variables que lo componen son las siguientes:
* **Time:** Número de segundos transcurridos entre la primer transacción del dataset (Time = 0) y la transacción en cuestión.
* **V1 a V28:** Variables originales transformadas y codificadas mediante la aplicación del modelo Principal Component Analysis.
* **Amount:** Monto de la transacción
* **Class:** Clasificador de fraude o no. Se trata de nuestro Target, una variable binaria, cuyo valor 1 equivale a una transacción fraudulenta y 0 a una transacción regular.

El link de Kaggle al post original del dataset es el siguiente: https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud/data

## Caso a analizar:

En la aplicación de un modelo de detección de fraude para transacciones en tarjetas de crédito se incluyen varios aspectos a tener en cuenta a la hora de desarrollar modelos de Machine Learning, tales como pueden ser el desbalanceo de los datos, los objetivos particulares de negocio del análisis o la capacidad de investigación sobre las alertas de detección que tenga la organización en cuestión. Todos estos conceptos deberán ser tenidos en cuenta no solo en el análisis, sino también en la implementación técnica, ya sea a la hora de explorar los datos como en el modelado y medición de resultados.

Con respecto a nuestro análisis, al tratarse de un modelo genérico y anonimizado, no contamos con directrices o información específica de un área de negocios, por lo que llevaremos adelante una serie de supuestos para determinar los lineamientos a seguir en cada fase del proyecto. Todos ellos serán explicados a continuación, intentando acercarnos mediante la lógica de negocio lo más posible a la realidad de una empresa que busca aplicar un modelo de detección de fraudes, con los beneficios y riesgos que esto implica, pero tratando de mejorar una situación inicial prácticamente sin modelo o con alguna regla muy simple (en nuestro caso, lo que llamaremos "Modelo Base").

Nuestro caso buscará predecir comportamientos fraudulentos de una mejor manera que la que lo harían modelos como series de reglas por ejemplo, los cuales podrán reemplazarse o funcionar de manera conjunta con nuestro modelo de Machine Learning, pero la ventaja que tendremos al aplicar este tipo de modelos, además de la automatización y dinamismo del mismo, será sobre todo el poder predecir patrones de conductas fraudulentas más complejos en donde actúan relaciones entre muchas más variables que las que un humano manualmente podría diseñar.

Normalmente este tipo de fraudes en transacciones de tarjetas de crédito pueden comenzar cuando un agente malintencionado utiliza identidades falsas, robadas o sintéticas, a menudo mediante el uso de métodos informáticos. Es por esto que <u>el modelo no solo busca determinar quien realizó un fraude y quien no, sino también entender porqué es más probable que un determinado tipo de transacciones sean más propensas a estar categorizadas como potenciales fraudes</u> y poder aplicar a tiempo los límites y las técnicas de investigación adecuadas para el caso.

Es por este motivo que nuestro objetivo no será solo determinar fraude o no fraude, sino que la **potencialidad de fraude** (expresado en probabilidades) tomará un rol importante también, lo que por consecuencia lleva a que las mediciones de los resultados del modelo no serán categóricas en base a una sola métrica, ni estarán orientadas solo a medir que tan bien el modelo percibe quien cometió fraude y quien no. De ser ese el caso, la medida a utilizar sería el clasico *Accuracy*, métrica que indica la cantidad de aciertos sobre el total de los datos predecidos, pero en nuestro caso, dotaremos de una mayor importancia a medidas como lo son el *Recall* y la *Precision*, junto a los que las combinan, el *F1* y el *F2 score*, además de por supuesto a la clásica *matriz de confusión*. Vamos a ahondar un poco en esto a continuación, ya que es un tema importante a tener en cuenta.

Si somos coherentes con lo indicado, nos será más importante, debido a que representará un beneficio mayor vs. el modelo base, el poder detectar la mayor cantidad de potenciales casos fraudulentos. Esto implica por supuesto el riesgo de obtener *falsos negativos* (casos de fraude no detectados por el modelo) y *falsos positivos* (casos que no son fraudulentos pero que el modelo determina que si pueden serlo), pero los objetivos de negocio justifican el correr este riesgo debido a la gran ganancia potencial al detectar correctamente nuevos fraudes y por ende detectar lo que más buscamos: los **verdaderos positivos**.

Siguiendo esta lógica, las métricas mencionadas tendran vital importancia y en nuestro caso sobre todo he decidido dotar de la mayor importancia relativa al **F2 Score**, una métrica que conjunta precision y recall pero que, a diferencia del F1 score, otorga mayor preponderancia al Recall. Es decir, que con esta decisión <u>se buscará premiar más a los modelos que, manteniendo un accuracy y una precision aceptables, permitan detectar la mayor cantidad posible de potenciales fraudes de manera coherente</u>, que luego serán analizados con mayor exhaustividad. Esto no quiere decir que solo me importe el Recall, debido a que también puede ser peligroso y muy costoso para una empresa tener que investigar demasiados casos por falta de precisión, por ello se elije el F2 por sobre simplemente el Recall. Esto desemboca en otro supuesto de negocio que deberemos realizar: la capacidad de investigación.

Al no contar con información de una empresa, no tenemos un número definido, pero al tratarse de modelos orientados a organizaciones de tipo bancario, que no solo consideramos que tendrían capacidad económica para financiar procesos de investigación, sino que además se trata de un área muy sensible del negocio tanto económica como reputacionalmente hablando, <u>supondremos que se cuenta con cierta capacidad de investigación, por lo que será aceptable el tener más casos a investigar (corriendo el riesgo de caer en falsos positivos) pero buscando detectar más fraudes, que tener muy pocos casos por más que estemos seguros que serán fraudulentos</u>. En caso de tener números concretos, esto podría llevarse a la práctica de manera efectiva agregando una matriz de costos al análisis. En este caso no podemos contar con ella, pero parece interesante al menos tratar de mostrar un ejemplo de como la aplicaría para tener una mejor visión numérica de los objetivos de negocio del modelo.

Además de lo comentado, será importante tener en cuenta que los datos, como suele suceder en casos de este estilo, tendán un nivel de **desbalanceo** muy importante, es decir que los fraudes serán muchos menos casos que los regulares, lo cual por supuesto tendrá consecuencias, complicando tanto en el análisis como en el modelado y por ende requiriendo técnicas específicas para el tratamiento de set de datos con este tipo de sesgos tan marcados, como lo son el **oversampling y undersampling**.

Finalmente, he de hacer una mención especial al apartado de la **explicabilidad** de los modelos que aplicaremos. Este concepto cada vez toma mayor relevancia en el mundo del data science en general, pero debemos tenerlo aún más en cuenta en casos del mundo de las finanzas, banca y aseguradoras debido no solo a la necesidad de comprensión de los modelos sino también a la creciente regulación gubernamental relativa a estos temas sensibles. Es por este motivo que realizaré un apartado especial referente a esta temática, intentando explicar de que manera el modelo toma las decisiones que toma para determinado tipo de comportamientos de los individuos.

He de matizar que no hablo de interpretabilidad y si de explicabilidad, debido a que dependiendo del modelo que elija, será posible ahondar más o menos en la precisión del análisis. Es decir, existen modelos que permiten tener un mayor entendimiento de su funcionamiento interno y determinar de manera prácticamente exacta de que manera afecta el valor de cada variable analizada al resultado final, como pueden ser por ejemplo modelos de regresión o de reglas, mientras que otro tipo de modelos que tienden más a clasificaciones complejas son más difíciles de interpretar de manera tan directa, como pueden ser las redes neuronales o los modelos de tipo bagging o boosting. Aún así, aplicando cierto tipo de librerías específicas para estos análisis, podremos aproximar una buena explicación en cuanto a cómo influyen las variables en las decisiones tomadas.

## Análisis sobre la variable Target

```python
df_fraud['Class'].value_counts(normalize=True)
# 0 es NO fraude y 1 si
```

```python
df_fraud['Class'].value_counts()
```

La variable 'Class' es el target que nos indica si se produjo fraude o no en la transacción

Como se puede notar, está claramente **desbalanceada**, hay solo 492 (un 0.17%) de transacciones fraudulentas

```python
target_count = df_fraud.groupby('Class').agg({'Class':'count'}).reset_index(drop=True)
target_count['value'] = list(target_count.index)
target_count
```

```python
fig, ax = plt.subplots()
sns.barplot(data=target_count, x='value', y='Class', ax=ax, hue='value', legend=False)
ax.set_title('Conteo de valores de la variable Target (Class)', fontdict={'fontsize':15}) #,'fontname':'Montserrat','weight':'bold'
ax.set_ylabel('Count')
ax.set_xlabel('Value')
for i in ax.containers:
    ax.bar_label(i,fmt='{:,.0f}')
plt.show()
```
