## Técnicas de reducción de dimensionalidad

En principio NO sería correcto aplicar este tipo de técnicas por 2 motivos:
* Principalmente debido a que NO se aprecia una correlación importante entre las variables
* Ya existe un PCA anterior, por más que no haya reducido dimensiones

Aún así, se va a demostrar que el modelo NO tiene un punto en el que mejore sustancialmente la explicación de la varianza

```python
ve_list = []
vne_list = []
n_comp = range(2,10)

for n in n_comp:
    pca = PCA(n_components=n)
    pca.fit(X_train_scaled.reset_index(drop=True))
    ve = pca.explained_variance_ratio_.sum()
    ve_list.append(ve)
    vne = 1-ve
    vne_list.append(vne)
    print(f'Varianza explicada por {n} components: {ve:.5f}')
    
plt.figure(figsize=(6, 6))
plt.plot(n_comp, vne_list, '-o', color='deepskyblue')
plt.xlabel(f'Número de componentes')
plt.ylabel('Varianza sin explicar')
plt.show()
```

```python
%%time
clf = RandomForestClassifier(n_estimators=200
                             ,criterion='gini'
                             ,max_depth=8
                             ,random_state=seed)
clf.fit(X, y)
```

```python
feature_importance = clf.feature_importances_ # obtengo la importancia de cada feature
feature_importance = 100.0 * (feature_importance / feature_importance.max()) # normalizo
sorted_idx = np.argsort(feature_importance) # ordeno los indices de menor a mayor
pos = np.arange(sorted_idx.shape[0]) + .5 # creo un array de posiciones
# plt.subplot(1, 2, 2)
plt.figure(figsize=(12, 20)) # defino el tamaño del gráfico
plt.barh(pos, feature_importance[sorted_idx], align='center') # creo el gráfico de barras
plt.yticks(pos, X.keys()[sorted_idx]) # defino los nombres de las features en el eje y
plt.xlabel('Relative Importance') # defino el nombre del eje x
plt.title('Variable Importance') # defino el título del gráfico
plt.show()
```

```python
# Variables menos importantes según Random Forest:
pd.concat([\
           pd.Series(X.keys()[sorted_idx].values)\
          ,pd.Series(feature_importance[sorted_idx])]\
          ,axis=1).head(15).rename(columns={0:'Variables',1:'Coef importancia'})
```

# Genero un pipeline en el que se encapsulen todos los tratamientos

Una vez mostrados todos los procesamientos, procederé a crear un pipeline que utilizaré en los siguientes notebooks para realizar el preprocessing de manera secuencial. Además ese pipeline será introducido en una función que irá en mi módulo de funciones .py, ubicado en la carpeta src, el cual recomiendo revisar para conocer las funciones que he creado y que utilizo a lo largo de los notebooks.

La idea es dejar listo un pipeline en donde se realicen secuencialmente todos los pasos del procesamiento, para que cuando ingrese el test o cualquier otro dataset, solo tenga que invocarlo y aplicar ese procesamiento.

Primero llevé a cabo una función simple de feature selection para invocarla rápidamente también:

```python
def feature_selection(df, add=[]):
    """
    ----------------------------------------------------------------------------------------------------------
    Función feature_selection:
    ----------------------------------------------------------------------------------------------------------
    - Funcionamiento: Recibe un DataFrame y un opcional de columnas extras a eliminar. Devuelve un DataFrame
    con las columnas del feature selection eliminadas, además de la columna extra en caso de haberse introducido.
    - Inputs:
        - df: DataFrame de Pandas al que se le reducirá el número de variables
        - add: argumento opcional en el que se pueden incluir más variables que se quieran eliminar
    - Return: DataFrame de pandas con las columnas reducidas según el feature selection aplicado.
    """
    drop = ['V23','V13','V25','V24', 'V19','V22']
    if add != []:
        drop+=add
    df_new = df.drop(drop, axis=1) #'device_fraud_count',
    return df_new
```
